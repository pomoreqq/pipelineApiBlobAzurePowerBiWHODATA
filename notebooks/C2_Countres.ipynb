```python
import pandas as pd 
import requests
from jsonschema import validate
import json
import fastparquet
```

```python
url = 'https://ghoapi.azureedge.net/api/DIMENSION/COUNTRY/DimensionValues'
```

```python
data = requests.get(url)
```

```python
jsonSchema = {
    "type": "object",
    "properties": {
        "Title": {"type:":"string"},
        "ParentDimension": {"type:":"string"},
        "Dimension": {"type:":"string"},
        "ParentCode": {"type:":"string"},
        "ParentTitle": {"type:":"string"}
    }
}
```

```python
dataJson = data.json()
dataJson['value']
```

```python
for i in dataJson['value']:
    validate(instance=i,schema=jsonSchema)
```

```python
for i in dataJson['value']:
    print(i)
```

```python

eu_countries = [
    "Austria",
    "Belgium",
    "Bulgaria",
    "Croatia",
    "Cyprus",
    "Czechia",
    "Denmark",
    "Estonia",
    "Finland",
    "France",
    "Germany",
    "Greece",
    "Hungary",
    "Ireland",
    "Italy",
    "Latvia",
    "Lithuania",
    "Luxembourg",
    "Malta",
    "Netherlands (Kingdom of the)",
    "Poland",
    "Portugal",
    "Romania",
    "Slovakia",
    "Slovenia",
    "Spain",
    "Sweden"
]
```

```python
euCountryList = []
for i in dataJson['value']:
    if i['Title'] in eu_countries:
        euCountryList.append(i)
```

```python
euCountriesDataFrame = pd.DataFrame(euCountryList)
euCountriesDataFrame
```
