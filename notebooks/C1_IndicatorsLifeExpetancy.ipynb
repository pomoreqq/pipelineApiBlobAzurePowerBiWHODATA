```python
import pandas as pd 
import requests
from jsonschema import validate
import json
import fastparquet
```

```python
baseURL= 'https://ghoapi.azureedge.net/api/Indicator'
```

```python
ListOfIndicatorCodesAndNamesLifeExpextancy = []
```

```python
def getIndicator(baseurl,nameOfIndicator):
    response = requests.get(baseurl+f"?$filter=IndicatorName eq '{nameOfIndicator}'")
    data = response.json()
    return data
```



1 Mortality and global health estimates

```python
firstIndicator = getIndicator(baseURL,'Adult mortality rate (probability of dying between 15 and 60 years per 1000 population)')
ListOfIndicatorCodesAndNamesLifeExpextancy.append(firstIndicator['value'])
ListOfIndicatorCodesAndNamesLifeExpextancy
```

```python
secondIndicator = getIndicator(baseURL,'Healthy life expectancy (HALE) at birth (years)')
ListOfIndicatorCodesAndNamesLifeExpextancy.append(secondIndicator['value'])
ListOfIndicatorCodesAndNamesLifeExpextancy
```

```python
thirdIndicator = getIndicator(baseURL,'Life expectancy at birth (years)')
ListOfIndicatorCodesAndNamesLifeExpextancy.append(thirdIndicator['value'])
ListOfIndicatorCodesAndNamesLifeExpextancy
```
