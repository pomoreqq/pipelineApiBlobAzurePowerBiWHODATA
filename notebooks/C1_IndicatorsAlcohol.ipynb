```python
import pandas as pd 
import requests
from jsonschema import validate
import json
import fastparquet
```

```python
baseURL= 'https://ghoapi.azureedge.net/api/Indicator'
```

```python
ListOfIndicatorCodesAndNamesAlcohol = []
```

```python
def getIndicator(baseurl,nameOfIndicator):
    response = requests.get(baseurl+f"?$filter=IndicatorName eq '{nameOfIndicator}'")
    data = response.json()
    return data
```

1️⃣ Alcohol Control Policies


2️⃣ Alcohol levels of consumption


3️⃣ Harms and consequences of alcohol


1.
Tax on alcoholic beverages: excise tax


Alcohol advertising restrictions on national television


Alcohol age limits, off-premise sales



```python
firstIndicator = getIndicator(baseURL,'Excise tax on alcoholic beverages')
ListOfIndicatorCodesAndNamesAlcohol.append(firstIndicator['value'][0])
ListOfIndicatorCodesAndNamesAlcohol
```

```python
SecondIndicator = getIndicator(baseURL,'Advertising restrictions on national television')
ListOfIndicatorCodesAndNamesAlcohol.append(SecondIndicator['value'][0])
ListOfIndicatorCodesAndNamesAlcohol
```

```python
ThirdIndicator = getIndicator(baseURL,'Age limits off-premise sales')
ListOfIndicatorCodesAndNamesAlcohol.append(ThirdIndicator['value'][0])
ListOfIndicatorCodesAndNamesAlcohol
```
