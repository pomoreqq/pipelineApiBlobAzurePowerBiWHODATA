```python
import pandas as pd 
import requests
from jsonschema import validate
import json
import fastparquet
```

```python
baseURL= 'https://ghoapi.azureedge.net/api/Indicator'
```

```python
ListOfIndicatorCodesAndNamesNCD = []
```

```python
def getIndicator(baseurl,nameOfIndicator):
    response = requests.get(baseurl+f"?$filter=IndicatorName eq '{nameOfIndicator}'")
    data = response.json()
    return data
```

1. NCD: Risk factors and conditions
2. NCD: Mortality

1.
Cholesterol, mean total (age-standardized estimate)
Hypertension among adults aged 30â€“79 years (%)
Diabetes, prevalence (%)

```python
firstIndicator = getIndicator(baseURL,'Mean total cholesterol,  age-standardized')
ListOfIndicatorCodesAndNamesNCD.append(firstIndicator['value'])
ListOfIndicatorCodesAndNamesNCD
```

```python
secondIndicator = getIndicator(baseURL,'Hypertension among adults aged 30-79 years, prevalence, age-standardized')
ListOfIndicatorCodesAndNamesNCD.append(secondIndicator['value'])
ListOfIndicatorCodesAndNamesNCD
```

```python
thirdIndicator = getIndicator(baseURL,'Prevalence of diabetes, age-standardized')
ListOfIndicatorCodesAndNamesNCD.append(thirdIndicator['value'])
ListOfIndicatorCodesAndNamesNCD
```

2.
Noncommunicable disease (NCD) mortality rate (per 100 000 population), age-standardized
 Noncommunicable disease (NCD) deaths, number
 Dying between the exact ages 30 and 70 years from cardiovascular diseases, cancer, diabetes, or chronic respiratory diseases, probability (SDG 3.4.1)

```python
fourthIndicator = getIndicator(baseURL,'Age-standardized NCD mortality rate  (per 100 000 population)')
ListOfIndicatorCodesAndNamesNCD.append(fourthIndicator['value'])
ListOfIndicatorCodesAndNamesNCD
```

```python
fifthIndicator = getIndicator(baseURL,'Total NCD Deaths')
ListOfIndicatorCodesAndNamesNCD.append(fifthIndicator['value'])
ListOfIndicatorCodesAndNamesNCD
```

```python
sixthIndicator = getIndicator(baseURL,'Probability (%) of dying between age 30 and exact age 70 from any of cardiovascular disease, cancer, diabetes, or chronic respiratory disease')
ListOfIndicatorCodesAndNamesNCD.append(sixthIndicator['value'])
ListOfIndicatorCodesAndNamesNCD
```

```python
flattenListOfIndicators = [x for innerlist in ListOfIndicatorCodesAndNamesNCD for x in innerlist]
dfIndicators = pd.DataFrame(flattenListOfIndicators)
dfIndicators
```

```python
categories = ['NCD: Risk factors and conditions',
              'NCD: Risk factors and conditions',
              'NCD: Risk factors and conditions',
              'NCD: Mortality',
              'NCD: Mortality',
              'NCD: Mortality']
dfIndicators['Category'] = categories
dfIndicators
```
